```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the Excel File
df = pd.read_csv('18+ Economics - Sheet.csv')

# Inspect columns
print(df.columns)

# Remove spaces or rename if necessary
df.columns = df.columns.str.strip()

# Now select the relevant columns
df_clean = df[["Income Level", "Reported registered", "Reported not registered"]]
df_clean["Reported registered"] = df_clean["Reported registered"].str.replace(",", "").astype(int)
df_clean["Reported not registered"] = df_clean["Reported not registered"].str.replace(",", "").astype(int)

# Continue with the melt and plotting steps as before...

# Step 3: Melt the Data for Grouped Bar Chart
df_melted = df_clean.melt(id_vars="Income Level", 
                          var_name="Registration Status", 
                          value_name="Count")

# Step 4: Plot Grouped Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Income Level", y="Count", hue="Registration Status", data=df_melted)

# Customize the Plot
plt.title("Registration Status by Income Level", fontsize=16)
plt.xlabel("Income Level", fontsize=12)
plt.ylabel("Number of People", fontsize=12)
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability
plt.legend(title="Registration Status")
plt.tight_layout()
plt.show()
```
