Below: two plots that analyze voter registration differences between 2020 and 2022. Trend: less voter registration overall in 2022 (makes sense, since it was the midterms)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

def reshape_data(df, year):
    # Extract status columns and add year
    df_long = pd.melt(df.reset_index(), id_vars=["index"], 
                      value_vars=["registered (percent)", "not registered (percent)", "no response (percent)"],
                      var_name="Status", value_name="Percent")
    df_long["Year"] = year
    return df_long.rename(columns={"index": "Race"})

# plot for native born
NB_data_2020 = reshape_data(NN2020nativeborndf, 2020)
NB_data_2022 = reshape_data(NN2022nativeborndf, 2022)
NB_combined_data = pd.concat([NB_data_2020, NB_data_2022])
NB_filtered_data = NB_combined_data[NB_combined_data["Status"] == "registered (percent)"]
plt.figure(figsize=(14, 8))
sns.barplot(
    data=NB_filtered_data,
    x="Race",  # Race on x-axis
    y="Percent",  # Percent on y-axis
    hue="Year",  # Color-coded by year
    ci=None,  # Disable confidence intervals for clarity
    palette="Set1"  # Use a color palette for differentiation
)
plt.title("Voter Registration by Race (Native Born) - 2020 vs 2022", fontsize=16)
plt.xlabel("Race", fontsize=14)
plt.ylabel("Percentage Registered", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Year", fontsize=12, title_fontsize=14)
plt.tight_layout()
plt.show()

# plot for naturalized
N_data_2020 = reshape_data(NN2020naturalizeddf, 2020)
N_data_2022 = reshape_data(NN2022naturalizeddf, 2022)
N_combined_data = pd.concat([N_data_2020, N_data_2022])
N_filtered_data = N_combined_data[N_combined_data["Status"] == "registered (percent)"]
plt.figure(figsize=(14, 8))
sns.barplot(
    data=N_filtered_data,
    x="Race",  # Race on x-axis
    y="Percent",  # Percent on y-axis
    hue="Year",  # Color-coded by year
    ci=None,  # Disable confidence intervals for clarity
    palette="Set1"  # Use a color palette for differentiation
)
plt.title("Voter Registration by Race (Naturalized) - 2020 vs 2022", fontsize=16)
plt.xlabel("Race", fontsize=14)
plt.ylabel("Percentage Registered", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Year", fontsize=12, title_fontsize=14)
plt.tight_layout()
plt.show()
```

Below: analysis of native born vs. naturalized for both years. It doesn't always seem to be the case that naturalized citizens have a higher rate of registration than native born ones. 

```python
def reshape_data_status_with_error(df, status): 
    df_long = pd.melt(
        df.reset_index(), 
        id_vars=["index"], 
        value_vars=["registered (percent)", "not registered (percent)", "no response (percent)"],
        var_name="Category", 
        value_name="Percent"
    ) 
    # margin of error
    moe_columns = pd.melt(
        df.reset_index(), 
        id_vars=["index"], 
        value_vars=["MOE 2", "MOE 4", "MOE 6"],  # Adjust if MOE column names differ
        var_name="Category_MOE", 
        value_name="MOE"
    )
    if len(df_long) != len(moe_columns):
        raise ValueError("Mismatch in lengths between Percent and MOE data")
    df_long["MOE"] = moe_columns["MOE"]
    df_long["Status"] = status
    return df_long.rename(columns={"index": "Race"})

# plot for 2020
NB_data_2020 = reshape_data_status_with_error(NN2020nativeborndf, "Native Born")
NT_data_2020 = reshape_data_status_with_error(NN2020naturalizeddf, "Naturalized")
combined_data_1 = pd.concat([NB_data_2020, NT_data_2020])
filtered_data_1 = combined_data_1[combined_data_1["Category"] == "registered (percent)"]
# creating plot
plt.figure(figsize=(14, 8))
sns.barplot(
    data=filtered_data_1,
    x="Race",  # Race on x-axis
    y="Percent",  # Percent on y-axis
    hue="Status",  # Color-coded by citizenship status
    ci=None,  # Disable automatic confidence intervals
    palette="Set2",  # Use a color palette for differentiation
)
# add error bars
for i, bar in enumerate(plt.gca().patches):
    error = filtered_data_1.iloc[i]["MOE"]
    bar_x = bar.get_x() + bar.get_width() / 2
    bar_height = bar.get_height()
    plt.errorbar(
        bar_x, bar_height, yerr=error, fmt="none", c="black", capsize=5, lw=1
    )
# customize plot
plt.title("Voter Registration by Race - Native Born vs. Naturalized (2020)", fontsize=16)
plt.xlabel("Race", fontsize=14)
plt.ylabel("Percentage Registered", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Citizenship Status", fontsize=12, title_fontsize=14)
plt.tight_layout()
plt.show()

# plot for 2022
NB_data_2022 = reshape_data_status_with_error(NN2022nativeborndf, "Native Born")
NT_data_2022 = reshape_data_status_with_error(NN2022naturalizeddf, "Naturalized")
combined_data_2 = pd.concat([NB_data_2022, NT_data_2022])
filtered_data_2 = combined_data_2[combined_data_2["Category"] == "registered (percent)"]
# creating plot
plt.figure(figsize=(14, 8))
sns.barplot(
    data=filtered_data_2,
    x="Race",  # Race on x-axis
    y="Percent",  # Percent on y-axis
    hue="Status",  # Color-coded by citizenship status
    ci=None,  # Disable automatic confidence intervals
    palette="Set2",  # Use a color palette for differentiation
)
# add error bars
for i, bar in enumerate(plt.gca().patches):
    error = filtered_data_2.iloc[i]["MOE"]
    bar_x = bar.get_x() + bar.get_width() / 2
    bar_height = bar.get_height()
    plt.errorbar(
        bar_x, bar_height, yerr=error, fmt="none", c="black", capsize=5, lw=1
    ) 
# customize plot
plt.title("Voter Registration by Race - Native Born vs. Naturalized (2022)", fontsize=16)
plt.xlabel("Race", fontsize=14)
plt.ylabel("Percentage Registered", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.legend(title="Citizenship Status", fontsize=12, title_fontsize=14)
plt.tight_layout()
plt.show()
```
