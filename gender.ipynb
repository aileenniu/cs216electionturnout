```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load in data
df = pd.read_csv("gender2022clean.csv")

# prep only the data columns needed (clean strings into proper numbers, remove rows)
df = df[df['Age'] != 'Total 18 years and over']
df['Reported registered citizen number'] = df['Reported registered citizen number'].str.replace(',', '')
df['Reported registered citizen number'] = pd.to_numeric(df['Reported registered citizen number'])

# preview dataframe
df
```

```python
# define plot palette:
customPalette = {'M': 'dodgerblue', 'F': 'violet'}

# make barplot
sns.catplot(
    data = df, 
    x = "Age", 
    y = "Reported registered citizen number", 
    hue = "Sex",
    palette = customPalette,
    kind = "bar"
)

# add titles and adjust
plt.title("Number of Registered Citizens by Age Category and Sex\nin the November 2022 Midterm Elections", fontsize = 14)

plt.ylabel("Number of Registered US Citizens (Thousands)", fontsize = 12)
plt.xlabel("Age Category", fontsize = 12)
plt.xticks(rotation=45, ha="right")  

# show the plot
plt.show()
```

```python
# define plot palette:
customPalette = {'M': 'dodgerblue', 'F': 'violet'}

# make barplot
sns.catplot(
    data = df, 
    x = "Age", 
    y = "Reported registered citizen percentage", 
    hue = "Sex",
    palette = customPalette,
    kind = "bar"
)

# add titles and adjust
plt.title("Percent of Registered Citizens by Age Category and Sex\nin the November 2022 Midterm Elections", fontsize = 14)

plt.ylabel("Percent of Registered US Citizens", fontsize = 12)
plt.xlabel("Age Category", fontsize = 12)
plt.xticks(rotation=45, ha="right")  

# show the plot
plt.show()
```
